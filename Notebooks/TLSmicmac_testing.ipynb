```python
folders = [f'/Volumes/ProjectsGarlockCSAF/CSAF_Lidar/TLS/MeeRanch/{sub}/MEC17_18/',
           f'/Volumes/ProjectsGarlockCSAF/CSAF_Lidar/TLS/MeeRanch/{sub}/MEC16_17/',
           f'/Volumes/ProjectsGarlockCSAF/CSAF_Lidar/TLS/MeeRanch/{sub}/MEC16_18/']
for fold in folders:
    im = gdal.Open(f'{fold}Correlmicmac.tif')
    im = im.GetRasterBand(1).ReadAsArray()
    mask = (im>=0.2)


    tt.projectDisp(f'{fold}EWmicmac.tif',
               f'{fold}NSmicmac.tif',-45,
               mask=mask,partif=f'{fold}ParallelDisp.tif',
               perptif=f'{fold}PerpendicularDisp.tif')
```
